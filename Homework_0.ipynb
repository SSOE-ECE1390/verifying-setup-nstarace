```python
# Import libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load spaceship image as grayscale (0 = color, 1 = grayscale)
ship = cv2.imread('Data/Apollo_11_Launch.jpg', 1)

# Plot image
plt.imshow(ship)
plt.axis('off') # Remove axis
plt.show()

```

```python
# Crop Image
h, w, c = ship.shape # First find dimensions
T = int(h/6) # Set bounds and round for integer pixels
B = int(5*h/6)
L = int(2*w/6)
R = int(4*w/6)
cropped_ship = ship[T:B, L:R] # Crop image

plt.imshow(cropped_ship)
plt.axis('off') # Remove axis
plt.show()
```

```python
# Brighten Image
bright_matrix = np.ones(cropped_ship.shape, dtype='uint8') * 80
bright_ship = cv2.add(cropped_ship, bright_matrix) # Use cv2.add to prevent overflow

plt.imshow(bright_ship)
plt.axis('off') # Remove axis
plt.show()
```

```python
# Rotate Image
h, w = bright_ship.shape[:2] # Get height and width
center = (w // 2, h // 2) # Find center
M = cv2.getRotationMatrix2D(center, 33, 1.0) # Get rotation matrix
rotated_ship = cv2.warpAffine(bright_ship, M, (w, h)) # Rotate image

plt.imshow(rotated_ship)
plt.axis('off') # Remove axis
plt.show()
```

```python
# Save Image
cv2.imwrite('Data/Edited_Apollo_11_Launch.jpg', rotated_ship)
```
